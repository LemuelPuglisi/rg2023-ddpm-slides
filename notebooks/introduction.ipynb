```python
import numpy as np
import skimage
import matplotlib.pyplot as plt
from skimage.io import imread, imsave
from skimage.exposure import rescale_intensity

im = imread('media/gorilla.jpg')
betas = np.linspace(0, 1, 5)

def diffuse(x_t, beta_t):
    eps = np.random.randn(*x_t.shape)
    return np.sqrt(1 - beta_t) * x_t + np.sqrt(beta_t) * eps, eps

def fastplot(x):
    plt.imshow(rescale_intensity(x, in_range=(-1, 1), out_range=(0, 1)))

fig, axes = plt.subplots(ncols=5, figsize=(20, 5))

x_t = im.copy() / 255.
x_t = rescale_intensity(x_t, in_range=(0, 1), out_range=(-1, 1))

images = []
noises = []

for i in range(5):
    ax = axes[i]
    ax.set_axis_off()
    x_next, eps = diffuse(x_t, betas[i])
    images.append(x_next)
    noises.append(eps)
    ax.imshow(rescale_intensity(x_next, in_range=(-1, 1), out_range=(0, 1)))
    x_t = x_next.copy()
    
plt.savefig('media/diffusion_process.png', facecolor=None)
plt.show()
```

```python
def save(x, fname):
    imsave(fname, rescale_intensity(x, in_range=(-1, 1), out_range=(0, 1)))
    
save(images[1], 'media/x_curr.png')
save(images[2], 'media/x_prev.png')
save(noises[1], 'media/eps_curr.png')
```

```python
for i, image in enumerate(images):
    save(image, f'media/image_{i}.png')
```
