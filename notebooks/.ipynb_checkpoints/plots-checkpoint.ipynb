```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.optimize import approx_fprime

sns.set(style="ticks", context="talk")
plt.style.use("dark_background")

GRID_Q = 35

def approx_score_function(f, shape, q):
    x, y = np.meshgrid(np.linspace(shape[0], shape[1], q), 
                       np.linspace(shape[0], shape[1], q))
    u = np.zeros(x.shape)
    v = np.zeros(y.shape)
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            s = (x[i,j], y[i,j])
            uv = approx_fprime(s, f)
            u[i,j], v[i,j] = uv
    return x, y, u, v

#----------------
mu1, cov1 = (5., 5.), 0.5 * np.eye(2)
X1 = np.random.multivariate_normal(mu1, cov1, 1000)

mu2, cov2 = (2., 2.), 0.3 * np.eye(2)
X2 = np.random.multivariate_normal(mu2, cov2, 1000)

#----------------
p1 = stats.multivariate_normal(mean=mu1, cov=0.5)
p2 = stats.multivariate_normal(mean=mu2, cov=0.1)
bp = lambda x: p1.pdf(x) + p2.pdf(x)

x, y, u, v = approx_score_function(bp, (0,8), GRID_Q)

sns.kdeplot(x=X1[:, 0], y=X1[:, 1], fill=True)
sns.kdeplot(x=X2[:, 0], y=X2[:, 1], fill=True)
plt.quiver(x,y, u, v, width=0.001)
plt.show()
```

```python
params = {'mathtext.default': 'regular'} 
plt.rcParams.update(params)

N_SAMPLES = 250
N_PLOTS = 5

fig, axes = plt.subplots(nrows=1, ncols=N_PLOTS, figsize=(20, 5))
axes = axes.flatten()

mu1, cov1 = (5., 5.), 0.5 * np.eye(2)
X1 = np.random.multivariate_normal(mu1, cov1, N_SAMPLES)

mu2, cov2 = (2., 2.), 0.3 * np.eye(2)
X2 = np.random.multivariate_normal(mu2, cov2, N_SAMPLES)

sigmas = np.linspace(0, 3, N_PLOTS)

for i in range(N_PLOTS):
    
    ax = axes[i]
    sigma = sigmas[i]
    ax.set_xlim(0, 8)
    ax.set_ylim(0, 8)
    if i > 0:
        ax.title.set_text("$p_{\sigma_" + str(i) + "}(x)$")
    else:
        ax.title.set_text("$p(x)$")
        
    ax.spines[['right', 'top']].set_visible(False)
    ax.set_facecolor('#121212')
    
    noise_1 = np.random.multivariate_normal([0, 0], sigma * np.eye(2), N_SAMPLES)
    noise_2 = np.random.multivariate_normal([0, 0], sigma * np.eye(2), N_SAMPLES)

    _X1 = X1 + noise_1
    _X2 = X2 + noise_2

    sns.kdeplot(x=_X1[:, 0], y=_X1[:, 1], fill=True, ax=ax, color=(0, 0.468, 0.738, 0.9))  
    sns.kdeplot(x=_X2[:, 0], y=_X2[:, 1], fill=True, ax=ax, color=(0.925, 0.378, 0.23, 0.9))

    X = np.concatenate([_X1, _X2])
    sns.scatterplot(x=X[:, 0], y=X[:, 1], ax=ax, color='#1C1C1C', linewidth=0, size=0.35, alpha=.3)
    ax.get_legend().remove()


plt.tight_layout()
plt.savefig('../public/adding_noise_score_matching.png', facecolor='#121212')
plt.show()
```
